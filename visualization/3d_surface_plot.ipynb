```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D



# ----  ----
mu = np.linspace(0, 0.8, 20)
sigma = np.linspace(0, 40, 20)
mu_grid, sigma_grid = np.meshgrid(mu, sigma)
NMI = np.exp(-(mu_grid**2 + (sigma_grid/40)**2))

# ----  ----
fig = plt.figure(figsize=(6, 5), dpi=300)
ax = fig.add_subplot(111, projection='3d') # 1   1  1 

surf = ax.plot_surface(
    mu_grid,
    sigma_grid,
    NMI,
    cmap='viridis',        # 
    edgecolor='k',        # 
    linewidth=0.3,
    antialiased=True,
    alpha=0.9            # 
)

# ----  X-Mark  ----
# ax.view_init(elev=35, azim=-70)

# ----  ----
ax.set_xlabel("", fontsize=12)
ax.set_ylabel("", fontsize=12)
ax.set_zlabel("NMI", fontsize=12)

# ----  ----
plt.tight_layout()
plt.show()
```
